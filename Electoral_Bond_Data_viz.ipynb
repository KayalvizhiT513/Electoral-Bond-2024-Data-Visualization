```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Path to the PDF file
excel_path = "/content/Details of Electoral Bonds submitted by SBI Part- I.xlsx"
```

```python
# Assuming your dataset is stored in a CSV file named 'electoral_bond_data.csv'
# Load the dataset
df = pd.read_excel(excel_path)

# Convert 'Date of Encashment' column to datetime format
df['Date of Encashment'] = pd.to_datetime(df['Date of Encashment'])

# Plot total amount of encashments over time
plt.figure(figsize=(10, 6))
df.groupby('Date of Encashment')['Denomination'].sum().plot()
plt.title('Total Amount of Encashments Over Time')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()

# Break down the encashments by political party over time
plt.figure(figsize=(12, 8))
for party, data in df.groupby('Name of the Political Party'):
    data.groupby('Date of Encashment')['Denomination'].sum().plot(label=party)
plt.title('Encashments by Political Party Over Time')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.legend()
plt.grid(True)
plt.show()

```
